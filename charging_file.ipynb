```python
df = df.dropna(how='any',axis=0)
```

```python
# get unique details
df.drop_duplicates(['Primary Used by'], inplace= True)
print(len(df))
```

```python
name = "SketchUPPro"
```

```python
lasel = "124063"
```

```python
df
```

```python
df1 = df.groupby(["RecCC"]).size().reset_index(name='counts')
```

```python
df1.counts.sum()
```

```python
list1 = list()
list2 = list()
```

```python
for i in df1["RecCC"]:
    list1.append(i[:3])
    list2.append(i[3:])
```

```python
df1["RecCC1"] = list1
df1["RecCC2"] = list2
```

```python
pd.set_option("display.max_rows", 500)
```

```python
counts = list()
```

```python
for i in df1["counts"]:
    counts.append(str(i)+"0000")
```

```python
df1["counts"] = counts
```
